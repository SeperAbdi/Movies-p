```python
from datascience import *
import numpy as np 
import matplotlib.pyplot as plots 
plots.style.use('fivethirtyeight')
%matplotlib inline
```

```python
table1=Table.read_table('top_movies.csv')
table1
```

```python
table2=table1.select('Title','Year')
table2
```

```python
sorted_by_Gross=table1.sort('Gross',descending=True)
sorted_by_Gross
```

```python
top=sorted_by_Gross.sort('Studio',distinct=True)
top
```

```python
top.select(1,2,3).sort(1,distinct=True).barh(0)
```

```python
ratio= table1.column("Gross (Adjusted)") / table1.column('Gross')
a = table1.with_column("Ratio",ratio)
a.sort('Ratio',descending=True).take(np.arange(10))
```
